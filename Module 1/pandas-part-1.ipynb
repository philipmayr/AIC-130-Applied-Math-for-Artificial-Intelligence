```python
# import pandas and numpy modules
import pandas as pd # data manipulation and analysis: https://pandas.pydata.org/docs/user_guide/index.html
import numpy as np  # scientific computing: https://numpy.org/doc/stable/user/index.html#user

# instantiate Pandas Series object: one-dimensional labeled array
numbers = pd.Series([1, 8, 7, 5, 6, 5, 3, 4, 7, 1])

# print original series
print(f"Original series:\n{numbers}")

# print list of indices of elements in original series whose immediate left- and right-hand neighbors (valleys) are both less than those elements (peaks, local maxima)
print("\nPositions of the values surrounded by smaller values on both sides:")

buffer = np.diff(np.sign(np.diff(numbers)))
result = np.where(buffer == -2)[0] + 1

print(result)
```
