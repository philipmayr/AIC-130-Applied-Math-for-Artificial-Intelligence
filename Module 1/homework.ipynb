```python
# https://matplotlib.org/stable/tutorials/pyplot.html
from matplotlib import pyplot as plt # import pyplot function to make matplotlib work like MATLAB

import numpy as np # https://numpy.org/doc/stable/
import pandas as pd # https://pandas.pydata.org/docs/

# generate 3x50 2D array (made up of 3 columns and 50 rows) filled in with random values drawn from standard normal distribution
x, y, scale = np.random.randn(3, 50)
# subplots() returns Figure and Axe objects - Figure is the whole sheet (or canvas) and Axe the drawing (or subplot)
fig, ax = plt.subplots() # subplots()

# create Series object - a 1D labeled array or list
s = pd.Series([1, 3, 5, np.nan, 6, 8]) # np.nan is a placeholder (not a number)
print(s)

# scatter() plots or "scatters" points at parameter coordinates x and y, c parameter sets point colors based on random values in scale variable 
ax.scatter(x=x, y=y, c=scale, s=np.abs(scale)*500) # last parameter s sets marker size (size of plotted points)
                                                   # np.abs(scale) finds absolute value of elements in scale array

ax.set(title="A Random Data Plot")

plt.show()
```
